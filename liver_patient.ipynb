```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/indian_liver_patient.csv')
df
```

```python
x = df.drop(columns=['Dataset'])
y = df['Dataset']
x
```

```python
x.isnull().sum()
#So there is no missing values
```

```python
xlatest = pd.get_dummies(x)
xlatest
```

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
ylatest = lr.fit_transform(y)
#totall = pd.DataFrame(data = ylatest, columns= y.columns)
ylatest
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
xlatest = minmax.fit_transform(xlatest)
#final = pd.DataFrame(final,ccolumns = df.columns)
xlatest
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(xlatest,ylatest,test_size=0.2,random_state=111)
xtest.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline


imputer = SimpleImputer(strategy='mean')


rf = RandomForestClassifier(random_state=42)


model = Pipeline([
    ('imputer', imputer),
    ('classifier', rf)
])


model.fit(xtrain, ytrain)


predictions = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
con = confusion_matrix(ytest,predictions)
con
```

```python
from sklearn.metrics import classification_report
cls = classification_report(ytest,predictions)
print(cls)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest,predictions)
acc
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline


imputer = SimpleImputer(strategy='mean')


dt_classifier = DecisionTreeClassifier(random_state=42)

model = Pipeline([
    ('imputer', imputer),
    ('classifier', dt_classifier)
])


model.fit(xtrain, ytrain)

predictions = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
con = confusion_matrix(ytest,predictions)
con
```

```python
from sklearn.metrics import classification_report
cls = classification_report(ytest,predictions)
print(cls)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest,predictions)
acc
```
